Análise de Modelo AR(1) para Previsão de Rentabilidade
Este notebook demonstra como calcular a previsão de longo prazo de uma série temporal modeled como um processo autoregressivo de ordem 1 (AR(1)). Especificamente, utiliza-se um modelo com um intercepto fixo ( \phi_0 = 8 ) e diferentes valores do parâmetro autoregressivo ( \phi_1 ).

Objetivos
Compreender como um modelo AR(1) permite estimar o valor de longo prazo da série.
Demonstrar a condição de estacionaridade e seu impacto na estabilidade do processo.
Visualizar como o valor de expectativa de longo prazo varia com diferentes valores de ( \phi_1 ).
Como usar
Preparação: Este notebook não requer instalação de pacotes adicionais além do matplotlib e numpy, que já estão disponíveis no Colab.

Execução: Execute as células para gerar os cálculos e o gráfico, que exemplificam a relação entre ( \phi_1 ) e o valor de longo prazo ( y^* ).

Interpretação: Observe o gráfico, que mostra como o valor de longo prazo depende de ( \phi_1 ). A análise ajuda a entender o comportamento de séries autoregressivas e identificar valores de ( \phi_1 ) que produzem previsões estacionárias ou não estacionárias.

```python
import matplotlib.pyplot as plt
import numpy as np

# Parâmetros
phi_0 = 8
phi_range = np.linspace(-1.5, 1.5, 300)  # range de phi_1, excluindo o ponto 1
y_long_term = []

# Calculando o valor de longo prazo para cada phi_1
for phi in phi_range:
    if abs(1 - phi) > 1e-5:  # evitando divisão por zero
        y_long_term.append(phi_0 / (1 - phi))
    else:
        y_long_term.append(np.nan)  # ponto não definido

# Plotando o gráfico
plt.figure(figsize=(10, 6))
plt.plot(phi_range, y_long_term, label='Valor de longo prazo $y^*$', color='blue')
plt.axhline(y=4, color='red', linestyle='--', label='Resposta (A): 4% (phi_1=-1)')
plt.scatter(-1, 4, color='red')  # marcar o ponto de interesse
plt.xlabel('$\phi_1$')
plt.ylabel('$y^*$ (%)')
plt.title('Valor de longo prazo $y^*$ em função de $\phi_1$')
plt.legend()
plt.grid(True)
plt.ylim(-10, 50)
plt.show()
```
